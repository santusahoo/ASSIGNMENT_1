```python
# question 01 / code 
#NAME : SANTU SAHOO | Reg .no: 22-27-20  |  Branch : Data science | Department : SoCE&MS
n=int(input("enter the no of element in list:: "))
list1=[]
for i in range (0,n):
   print("enter element for index:",i)
   list1.append(int(input()))
print("given list is:",list1)
list2 = list1.copy()
list1.reverse()
print("reverse list is:",list1)
list3 = []
for i in range (0 ,n):
    list3.append(list1[i]*list2[i])

print("resultane list is:: " ,list3)
```

```python
# question 05 / code 
#NAME : SANTU SAHOO | Reg .no: 22-27-20  |  Branch : Data science | Department : SoCE&MS
num=int(input("enter the no of element to find  factorial "))
def factorial(n):
  if (n==1 or n==0):
    return 1
  else:
    return (n*factorial(n-1))

result= factorial(num)
print(f"factorial {num}! =" , result)

  
```
